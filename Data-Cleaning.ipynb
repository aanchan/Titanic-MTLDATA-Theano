To discard or to not discard, that is the question
----
There are many places where information about a passenger is unavailable, e.g. 'Age' has many NaN values.
How do you deal with this incomplete information?

Use the Pandas DataFrame isnull() function, DataFrame logical indexing and DataFrame Field referencing 
to see some of the entries which have NaN/null age values

```python
#Field Referencing ---> df['Age']
#isNull Function ----> df['Age'].isnull()
#Logical Indexing ---> df[df['Age'].isnull()]
nullAgeDataFrame=df_train[df_train['Age'].isnull()]
#To view sub-fields of a frame, provide the subfields as a list ---> ['Sex', 'Pclass', 'Age']
nullAgeDataFrame[['Sex', 'Pclass', 'Age']]
```

A Solution : Replace null/Nan age entries with median age values. 
---
Median age values are to be calculated over Sex and PClass.
    * Sex : M,F
    * PClass : 1,2,3
    * Therefore, 6 median age values are to be calculated

An aside, replace Sex - F/M with numeric values 0/1
---
This would make data processing easier

```python
#create a new column called Gender and populate it with 1 : Male, and 0:Female 
#Training data 
df_train['Gender'] = df_train['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df_train.head()
```

```python
#create a new column called Gender and populate it with 1 : Male, and 0:Female 
#Test data
df_test['Gender'] = df_test['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df_test.head()
```

Now back to finding out the median age values over Sex, and PClass
----

```python
median_ages = np.zeros((2,3))#Rows correspond to females, males (2 Rows). Columns correspond to the class (3 Columns)
median_ages
```

```python
#Use a nested loop to calculate the medians
for i in range(0, 2):#loop over the Sex (Rows)
    for j in range(0, 3):#loop over the classes (Columns)
        #Here first short list all entries that correspond to the required Sex
        # ---> (df_train['Gender'] == i) 
        # And ---> &
        #In that short list, pick out the ones that correspond to a particular class
        #---->df_train['Pclass'] == j+1
        # Drop null age values using dropna() on a dataFrame and calulate the median
        #Store median for Sex i, and Class j in median(i,j)
        median_ages[i,j] = df_train[(df_train['Gender'] == i) & \
                              (df_train['Pclass'] == j+1)]['Age'].dropna().median()
        
median_ages
```

Create a new column called 'AgeFill' for train and test data frames
----

```python
#Training AgeFill
df_train['AgeFill'] = df_train['Age']
df_test['AgeFill'] = df_test['Age']
```

```python
for i in range(0, 2):#loop over the Sex (Rows)
    for j in range(0, 3):#loop over the classes (Columns)
        #Get the locations corresponding to Class and and Gender where the age is null and replace it
        #with the median age in our new field AgeFill
        df_train.loc[ (df_train.Age.isnull()) & (df_train.Gender == i) & (df_train.Pclass == j+1),'AgeFill'] = median_ages[i,j]
        
#Test AgeFill
for i in range(0, 2):#loop over the Sex (Rows)
    for j in range(0, 3):#loop over the classes (Columns)
        #Get the locations corresponding to Class and and Gender where the age is null and replace it
        #with the median age in our new field AgeFill
        df_test.loc[ (df_test.Age.isnull()) & (df_test.Gender == i) & (df_test.Pclass == j+1),'AgeFill'] = median_ages[i,j]
        
df_train.head()
```

Now for some feature engineering
---
    a. Create a new feature called Age*Class
    b. Create a new feature called FamilySize = Number of Siblings/Spouses + Number of Parents/Children on -board

```python
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch']
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch']

df_train['Age*Class'] = df_train.AgeFill * df_train.Pclass
df_test['Age*Class'] = df_test.AgeFill * df_test.Pclass

df_train.loc[888,:]
```
